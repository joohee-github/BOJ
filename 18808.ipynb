#### 2차원 회전배열 알고리즘

```python
# 90도 회전
'''
회전 전의 열과 후의 행이 일치한다.
따라서 두 곳에는 j값을 그대로 준다.
회전 후의 열은 n-1에서 회전 전의 행을 뺀 값이다.
회전 전의 행번호가 i일때 n-1에서 i값을 빼준다.
'''

def rotated(a):
    n = len(a)
    m = len(a[0])
    
    result = [[0] * n for _ in range(m)]
    
    for i in range(n):
        for j in range(m):
            result[j][n-i-1] = a[i][j]
        return result
```

```python
# 180도 회전
'''
회전 후의 행번호는 N-1값에서 전의 행 번호를 뺀 것과 같다.
회전 후의 열 번호는 N-1값에서 열 번호를 뺀 것과 같다.
'''
# 이 문제에서는 행과 열의 크기가 같다고 가정

def rotate_180(m):
    N = len(m)
    ret = [[0]*N for _ in range(N)]
    
    for r in range(N):
        for c in range(N):
            ret[N-1-r][N-1-c] = m[r][c]
    
    return ret
```

```python
# 270도 회전
'''
회전 후의 열과 전의 행이 일치한다.
후의 행번호는 N-1 에서 전의 열 번호를 뺀 값과 일치한다.
'''

def rotate_270(m):
    N = len(m)
    ret = [[0]*N for _ in range(N)]
    
    for r in range(N):
        for c in range(N):
            ret[N-1-c][r] = m[r][c]
    return ret
```
